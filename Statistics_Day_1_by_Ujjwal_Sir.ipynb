# Statistics

It is divided into 2 parts -

1) **Descriptive statistics** - ( Mathematical operations )

It is a branch of statistics that involves collection and interpretation of data .

Descriptive stats is also divided into 2 parts -

-> **Measure of Central Tendancy**  : It is used to indicate where does the centre of my data lies .

Mean

Median

Mode

-> **Measure of Variance / Dispersion** : It is used to indicate how much my data is spread out / Dispersed in all the direction .

Range

Variance

Standard Deviation

Percentile

Quartile

Outliers

Correlation


--------------------------------------------------------------------------

1) **Hypothesis Testing**

```python
import statistics as stats
import numpy as np
```

```python
# Mean ( Average of the data )

'''
data =[ 5,2,7,8,5,9,1,3,5]

I want to find average / mean of this data ?

Mean = Sum of all values / Total no of values
Mean = 45 / 9
Mean = 5
'''
data =[ 5,2,7,8,5,9,1,3,5]
ans = stats.mean(data)
print(ans)

data =[ 5,2,7,8,5,9,1,3,5]
ans = np.mean(data)
print(ans)
```

```python
# Median ( Middle value of your data )

'''
data =[5,2,7,8,5,9,1,3,5]

Median -  Before finding median / middle value  first sort your data

data =[ 1 , 2 , 3 , 5, 5, 5, 7 ,8 ,9 ]
Median = 5
'''
data =[5,2,7,8,5,9,1,3,5]
ans = stats.median(data)
print(ans)

data =[5,2,7,8,10,5,6,13,9,1,3,5]
ans = stats.median(data)
print(ans)

data =[5,2,7,8,10,5,6,13,9,1,3,5]
ans = np.median(data)
print(ans)
```

```python
# Mode ( Most common value / Most repeated value / Value having the highest frequency)

data = [ 1 ,4,11,12,23,23,23,56,78,1,1,23,4,11,19]

'''
What is the Mode of this data
1- 3 times
4- 2 times
11- 2 times
12- 1 times
19- 1 times
23- 4 times   <- Highest frequency
56- 1 times
78- 1 times

23 is the Mode of this data
'''
data = [ 1 ,4,11,12,23,23,23,56,78,1,1,23,4,11,19]
ans = stats.mode(data)
print(ans)
```

```python
data = [ 1 ,4,11,11,11,12,23,23,23,56,78,1,1,23,4,11,19]

'''
What is the Mode of this data
1- 3 times
4- 2 times
11- 4 times   <- Highest frequency
12- 1 times
19- 1 times
23- 4 times   <- Highest frequency
56- 1 times
78- 1 times

RULE - If more than one value is having the same highest frequency then ?
The value that will come first in your data is going to be the Mode !
'''
data = [ 1 ,4,11,11,11,12,23,23,23,56,78,1,1,23,4,11,19]
ans = stats.mode(data)
print(ans)

data = [ 1 ,4,23,11,11,11,12,23,23,23,56,78,1,1,4,11,19]
ans = stats.mode(data)
print(ans)
```

 **Measure of Variance / Dispersion**

```python
# Range ( Difference between Maximum ans Minimum Value )

'''
data =[ 25,30,35,45,40,60,20,65,70,10]
Range = Maximum value - Minimum Value

Range = 70 - 10
Range = 60
'''

# 1st option to find Range
data =[ 25,30,35,45,40,60,20,65,70,10]
max= np.max(data)
min= np.min(data)

print("Maximum value is :",max)
print("Minimum value is :",min)

Range = max - min
print("Range of the data is :",Range)
```

```python
# 2nd option to find Range
data =[ 25,30,35,45,40,60,20,65,70,10]
ans = np.ptp(data)                         # ptp - peak to peak
print(ans)
```

```python
# Variance - It quantifies the spread of your data around mean

'''
data =[ 25,30,35,40,45,55,50 ,60,65,70]
variance = Sum of squared difference between mean and each data points / Total no of data points

Mean of data = 47.5
variance =[(47.5 -25)^2 +(47.5 -30)^2 +(47.5 -35)^2+ (47.5 -40)^2 + (47.5 -45)^2 + (47.5 -55)^2 + (47.5 -50)^2 + (47.5 -60)^2 + (47.5 -65)^2 + (47.5 -70)^2] / 10
variance = [(-22.5)^2 +(-17.5)^2 +(-12.5)^2 +(-7.5)^2+(-2.5)^2+(7.5)^2+(2.5)^2+(12.5)^2+(17.5)^2+(22.5)^2] /10

# Ignore - and + sign as square will cancel them

variance =206.5
'''
data =[ 25,30,35,40,45,55,50 ,60,65,70]
ans=np.var(data)
print(ans)
```

Numpy library will use =

variance = Sum of squared difference between mean and each data points / Total no of data points

Stats library will use =

variance = Sum of squared difference between mean and each data points / Total no of data points -1

```python
# Variance - It quantifies the spread of your data


'''
# Height of indian man's  - The value in this data is close to each other
height =[ 5 , 5.5 , 6 , 6.2 , 5.9 , 5.6 ,5.2 , 5.4 , 5.8 ]

# Per day income of indian man's - The value in this data is far away to each other
income = [ 100 , 12000 , 500 , 33000 , 100000 , 5600000 , 8900000 , 50 , 600 ]

# Low variance of any data means value in the data are close to each other
# Hight variance of any data means value in the data are far away to each other

'''
height =[ 5 , 5.5 , 6 , 6.2 , 5.9 , 5.6 ,5.2 , 5.4 , 5.8 ]
ans=np.var(height)
print("Variance of height data :",ans)

income = [ 100 , 12000 , 500 , 33000 , 100000 , 5600000 , 8900000 , 50 , 600 ]
ans=np.var(income)
print("Variance of income data :",ans)
```

```python
data = [ 1,2,3,4,55,66,12000,242343,7445,32424,7563]
# High variane or low variance
ans=np.var(data)
print("Variance of income data :",ans)
```

```python
# Standard deviation = Square root of variance ( It indicates how much an individual data point deviated from mean)

# 1st Method
data =[ 25,30,35,40,45,55,50 ,60,65,70]
variance=np.var(data)  # Variance of data

standard_deviation = np.sqrt(variance)
print(standard_deviation)

# 2nd Method
data =[ 25,30,35,40,45,55,50 ,60,65,70]
standard_deviation = np.std(data)
print(standard_deviation)
```
